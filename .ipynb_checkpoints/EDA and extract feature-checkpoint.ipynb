```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read train and test file
train = pd.read_csv('input/train.csv',index_col='PassengerId')
test = pd.read_csv('input/test.csv',index_col='PassengerId')
```

```python
train.head(10)
```

```python
test.head(10)
```

```python
#Overview Data
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
sns.countplot('Survived',data=train,ax=axes[0,0])
sns.countplot('Pclass',data=train,ax=axes[0,1])
sns.countplot('Sex',data=train,ax=axes[0,2])
sns.countplot('SibSp',data=train,ax=axes[0,3])
sns.countplot('Parch',data=train,ax=axes[1,0])
sns.countplot('Embarked',data=train,ax=axes[1,1])
sns.distplot(train['Fare'], kde=True,ax=axes[1,2])
sns.distplot(train['Age'].dropna(),kde=True,ax=axes[1,3])
```

```python
sns.catplot(x='Sex', col='Survived', kind='count', data=train);
sns.catplot(x='Embarked', col='Survived', kind='count', data=train);
sns.catplot(x='Pclass', col='Survived', kind='count', data=train);
```

```python
datadict = pd.DataFrame(train.dtypes)
datadict['missing'] = train.isnull().sum()
datadict['NUnique'] = train.nunique()
datadict['count'] = train.count()
datadict = datadict.rename(columns={0:'DataType'})
```

```python
datadict
```

```python
figbi, axesbi = plt.subplots(2, 4, figsize=(16, 10))
train.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
train.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
train.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
train.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
train.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=train,ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Fare", data=train,ax=axesbi[1,2])
```

```python
sns.jointplot(x="Age", y="Fare", data=train);
```

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = train.corr()
sns.heatmap(corr,
            mask=np.zeros_like(corr, dtype=np.bool), 
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

```python
#finding some feature from string data
train['Embarked'] = train['Embarked'].fillna('S')
train['Embarked'] = train['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
train['isincabin'] = train.Cabin.isna()
train['isincabin'] = train['isincabin'].map( {True : 0, False: 1}).astype(int)
train['Sex'] = train['Sex'].map({'female' : 1, 'male': 0}).astype(int)
train['familysize'] = train.SibSp+train.Parch+1
train['title'] = train.Name.str.extract('\, ([A-Z][^ ]*\.)',expand=False)
train['Name_len']=train.Name.str.len()
train['Ticket_First']=train.Ticket.str[0]
```

```python
figbi, axesbi = plt.subplots(2, 2, figsize=(16, 10))
train.groupby('isincabin')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
train.groupby('familysize')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
```

```python
train.title.value_counts().reset_index()
```

```python
train.Age.fillna(train.Age.median(),inplace=True)
train.Fare.fillna(train.Fare.median(),inplace=True)
```

# Check test data and extract feature

```python
datadict_test = pd.DataFrame(test.dtypes)
datadict_test['missing'] = test.isnull().sum()
datadict_test['NUnique'] = test.nunique()
datadict_test['count'] = test.count()
datadict_test = datadict_test.rename(columns={0:'DataType'})
datadict_test
```

```python
test.Age.fillna(test.Age.median(),inplace=True)
test.Fare.fillna(test.Fare.median(),inplace=True)
#finding some feature from string data
test['isincabin'] = test.Cabin.isna()
test['isincabin'] = test['isincabin'].map( {True : 0, False: 1}).astype(int)
test['familysize'] = test.SibSp+test.Parch+1
test['title'] = test.Name.str.extract('\, ([A-Z][^ ]*\.)',expand=False)
test['Embarked'] = test['Embarked'].fillna('S')
test['Embarked'] = test['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
test['Sex'] = test['Sex'].map({'female' : 1, 'male': 0}).astype(int)
test['Name_len']=test.Name.str.len()
test['Ticket_First']=test.Ticket.str[0]
```

```python
train.to_csv('input/train_data.csv',index=True)
test.to_csv('input/test_data.csv',index=True)
```
