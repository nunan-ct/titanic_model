```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

```python
#Read train and test file
train = pd.read_csv('input/train_data.csv',index_col='PassengerId')
test = pd.read_csv('input/test_data.csv',index_col='PassengerId')
```

```python
train.head(10)
```

```python
train.columns
```

```python
#using train_col , train_y
#'SibSp', 'Parch' ,'Embarked'
usecol = ['Pclass','Sex', 'Age','Fare', 'isincabin', 'familysize','Name_len']
Xtrain = train[usecol]
Ytrain = train['Survived']
```

```python
model = RandomForestClassifier(n_estimators=100,random_state=123)
#n_estimators=80,learning_rate=0.01,max_depth=8,max_features='sqrt',min_samples_split=12
model.fit(Xtrain,Ytrain)
#print(model.score(Xtest,Ytest))
df_imp = pd.DataFrame({'name':Xtrain.columns,'imp':model.feature_importances_}).sort_values('imp',ascending=False)
df_imp.head(20)
```

```python
# Make a prediction
y_predict = model.predict(Xtrain)
y_predict[:10]
(Ytrain== y_predict).mean()
```

```python
Xtest = test[usecol]
testtest = model.predict(Xtest)
test['Survived'] = testtest
passenger_id = test['Survived']
passenger_id.to_csv('submission.csv',sep=',',index=True,index_label='PassengerId',header=True)
```
